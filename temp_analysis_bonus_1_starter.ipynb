### Compare June and December data across all years 

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

# Other Dependencies
from scipy import stats
```

```python
# Create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# Reflect an existing database into a new model
Base = automap_base()

# Reflect the tables
Base.prepare(engine, reflect = True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Identify the average temperature for June
june_avg_temp = session.query(func.avg(measurement.tobs)).\
    filter(measurement.date.like("%-06-%")).all()
june_avg_temp
```

```python
# Create collection of June temperature data
june_temps = session.query(measurement.tobs).\
    filter(measurement.date.like("%-06-%")).all()

june_temps_list = []
for x in june_temps:
    june_temps_list.append(x)

june_temps_list
```

```python
# Identify the average temperature for December
dec_avg_temp = session.query(func.avg(measurement.tobs)).\
    filter(measurement.date.like("%-12-%")).all()
dec_avg_temp
```

```python
# Create collection of December temperature data
dec_temps = session.query(measurement.tobs).\
    filter(measurement.date.like("%-12-%")).all()

dec_temps_list = []
for x in dec_temps:
    dec_temps_list.append(x)

dec_temps_list
```

```python
# Run paired t-test
stats.ttest_ind(june_temps_list, dec_temps_list)
```

```python
session.close()
```

### Analysis

An independent/unpaired t-test was used since we are not comparing one group against a standard value, rather we are comparing two distinct groups against each other. In this case, the p-value is extremely small and less than 0.05, thus the difference in means is statistically significant.
